```python
trained_teacher_ratio = pd.DataFrame()
fm_teacher_ratio = pd.DataFrame()
pupil_teacher_ratio = pd.DataFrame()
gross_enrol_ratio = pd.DataFrame()
girl_boy_ratio = pd.DataFrame()

year_name = {2008: '2008-09', 2009: '2009-10', 2010: '2010-11'}

def clean_state(state):
    try:
        state = state.rstrip()
        state = state.replace('\n', ' ')
        state = state.replace('&', 'and')
        return state.rstrip()
    except:
        return state

def state_group(state):
    d = {'Uttar Pradesh': 'North - Plains', 'Bihar': 'North - Plains', 'Madhya Pradesh': 'North - Plains', 
         'Rajasthan': 'North - Plains', 'Punjab': 'North - Plains', 'Haryana': 'North - Plains',
         'Sikkim': 'North-East', 'Manipur': 'North-East', 'Tripura': 'North-East', 'Nagaland': 'North-East',
         'Assam': 'North-East', 'Meghalaya': 'North-East', 'Arunachal Pradesh': 'North-East', 'Mizoram': 'North-East',
         'Gujarat': 'West', 'Maharashtra': 'West',
         'Chattisgarh': 'East', 'Jharkhand': 'East', 'Odisha': 'East', 'Orissa': 'East', 'West Bengal': 'East',
         'Chhattisgarh': 'East',
         'Karnataka': 'South', 'Andhra Pradesh': 'South', 'Kerala': 'South', 'Tamil Nadu': 'South',
         'Jammu & Kashmir': 'North - Mountains', 'Himachal Pradesh': 'North - Mountains', 'Uttarakhand': 'North - Mountains',
         'Jammu and Kashmir': 'North - Mountains', 'INDIA': 'India', 'India': 'India'}
    if state in d:
        return d[state]
    else:
        return 'Uncategorized'

for year in xrange(2008, 2011):
    xl = pd.ExcelFile('HRD Data/%s_edited.xlsx'%year_name[year])
    
    tt_ratio = xl.parse('TrainedTeacher')
    header = tt_ratio.columns
    header = [i.rstrip().replace('\n', ' ') for i in header]
    tt_ratio.columns = header
    tt_ratio['State'] = tt_ratio['State'].apply(clean_state)
    tt_ratio['Year'] = year
    tt_ratio['State Group'] = tt_ratio['State'].apply(state_group)
    trained_teacher_ratio = trained_teacher_ratio.append(tt_ratio)
    
    fmt = xl.parse('F-MTeacher')
    fmt['State'] = fmt['State'].apply(clean_state)
    fmt['Year'] = year
    fmt['State Group'] = fmt['State'].apply(state_group)
    header = fmt.columns
    header = [i.rstrip().replace('\n', ' ') for i in header]
    fmt.columns = header
    fm_teacher_ratio = fm_teacher_ratio.append(fmt)
    
    ptr = xl.parse('PTR')
    ptr['State'] = ptr['State'].apply(clean_state)
    ptr['Year'] = year
    ptr['State Group'] = ptr['State'].apply(state_group)
    header = ptr.columns
    header = [i.rstrip().replace('\n', ' ') for i in header]
    ptr.columns = header
    pupil_teacher_ratio = pupil_teacher_ratio.append(ptr)
    
    ger = xl.parse('GERAll')
    ger = ger.ix[1:]
    ger['State'] = ger['State'].apply(clean_state)
    ger['Year'] = year
    ger['State Group'] = ger['State'].apply(state_group)
    header = ger.columns
    header = [i.rstrip().replace('\n', ' ') for i in header]
    ger.columns = header
    gross_enrol_ratio = gross_enrol_ratio.append(ger)
    
    gbr = xl.parse('G-B')
    gbr['State'] = gbr['State'].apply(clean_state)
    gbr['Year'] = year
    gbr['State Group'] = gbr['State'].apply(state_group)
    header = gbr.columns
    header = [i.rstrip().replace('\n', ' ') for i in header]
    gbr.columns = header
    girl_boy_ratio = girl_boy_ratio.append(gbr)
```

**The ASER Reading test**
All children in the age group 5-16 are administered a “floor level” reading test in the language of their choice (the test is available in 16 Indian languages). The highest level tested is equivalent to a Std 2 level text. Each child is marked at the highest level at which s/he can read comfortably:

Story : Can read a longer paragraph (Std 2 level text)

Para : Can read a short paragraph (Std 1 level text)

Word : Can read 4 out of 5 words correctly

Letter : Can identify 4 out of 5 letters correctly

Nothing : Identifies fewer than 4 out of 5 letters correctly.


**The ASER Maths Test**

All children in the age group 5-16 are administered a “floor level” test of basic arithmetic. The highest level tested is 3-digit by 1-digit division. Each child is marked at the highest level which s/he can do comfortably:

Division : Can solve a 3-digit by 1-digit division problem

Subtraction : Can solve two 2-digit by 2-digit subtraction problems with carryover

Number recognition 11-99 : Can identify 4 out of 5 numbers between 11 and 99

Number recognition 1-9 : Can identify 4 out of 5 numbers between 1 and 9

Nothing : Identifies fewer than 4 out of 5 single-digit numbers correctly.

```python
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

def legend_map(s):
    d = {'East': 'East India', 'North - Plains': 'North Indian Plains', 'North - Mountains': 'North Indian Mountains',
         'South': 'South India', 'West': 'West India', 'North-East': 'North-Eastern India'}
    if s in d:
        return d[s]
    else:
        return s

def plot_gross_enrolment_compare(states, title, text=None, state_level = False):
    width = 1./(len(states)+1)
    for state in states:
        if not state_level:
            df = gross_enrol_ratio[gross_enrol_ratio['State Group'] == state].groupby('State Group').agg('mean')
        else:
            df = gross_enrol_ratio[gross_enrol_ratio['State'] == state].groupby('State Group').agg('mean')
        cat = ['Classes I-V', 'Classes VI-VIII', 'Classes IX-X', 'Classes XI-XII']
        sub_cat = ['Boys', 'Girls', 'Total']

        plot_pos = 0
        boys_vals = []
        girls_vals = []
        total_vals = []

        for i in xrange(len(cat)):
            for j in xrange(len(sub_cat)):
                val = df.values[0][3*i+1+j]
                if j == 0:
                    boys_vals.append((val,plot_pos+0.25))
                elif j == 1:
                    girls_vals.append((val,plot_pos+0.25))
                else:
                    total_vals.append((val,plot_pos-0.5))
                plot_pos += 0.25
                plt.bar
            plot_pos += 0.25
        boys_vals, boys_ind = [list(t) for t in zip(*boys_vals)]
        girls_vals, girls_ind = [list(t) for t in zip(*girls_vals)]
        total_vals, total_ind = [list(t) for t in zip(*total_vals)]
        plt.bar([i + states.index(state)*width for i in xrange(len(total_ind))], total_vals, width, label = 'Total', color = dark2_colors[states.index(state)])
        #plt.bar(total_ind, total_vals, 0.25, label = 'Total', color = dark2_colors[5])
        #plt.bar(boys_ind, boys_vals, 0.25, label = 'Boys', color = dark2_colors[2])
        #plt.bar(girls_ind, girls_vals, 0.25, label = 'Girls', color = dark2_colors[3])
    plt.xticks([i + 0.4 for i in xrange(4)], cat)
    plt.title(title)
    plt.gca().grid(True)
    plt.ylim(0,max(total_vals)+10)
    if not state_level:
        states = [legend_map(i) for i in states]
    plt.legend(states)
    plt.text(0, -15, 'Source: Ministry of Human Resource Development (http://mhrd.gov.in/statist?field_statistics_category_tid=33)', fontsize = 10)
    plt.text(0, -20, 'Note: These can exceed 100% due to the inclusion of over-aged and under-aged students because of early or late', fontsize = 10)
    plt.text(0, -26, 'school entrance and grade repetition.\nTheNumerateIndian.com', fontsize = 10)
    if text:
        for t in text:
            plt.text(0, -31 - 3*text.index(t), t, fontsize = 10)
    plt.show()
```
