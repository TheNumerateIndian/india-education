```python
states = df_gender_read['state'].unique()
year_wise = {}
median_change = {}
for s in states[:]:
    year_wise[s] = df_gender_read[df_gender_read.reading_level.isin(['Story', 'Std2_Para'])][
                        df_gender_read.gender == 'All'][df_gender_read.state == s][['year', 'percent']]
    year_wise[s] = year_wise[s].set_index('year', drop=True)
    year_wise[s]['change'] = pd.Series(index=year_wise[s].index)
    for y in range(2007, 2014):
        if y in year_wise[s].index and (y-1) in year_wise[s].index:
            year_wise[s]['change'][y] = year_wise[s].percent[y] - year_wise[s].percent[y-1]
    median_change[s] = year_wise[s].change.median()
#    fig, ax = plt.subplots()
#    sub_plot = year_wise[s][['change']].plot(kind='bar', ax=ax)
#    sub_plot.set_title(s)
    #sub_plot.set_ylim([0,10])
                                 
```
